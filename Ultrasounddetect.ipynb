```python
# For Data augmentation i resorted to use 5% to 10% rotational different images, also used a blur factor of 2.5px because in ultrasound images, rotational variations (5%-10%) mimic real-world probe positioning also introducing a blur factor (2.5px) simulates image clarity fluctuations, boosting realism. Both augmentations help the model generalize to diverse ultrasound scan scenarios, improving accuracy and robustness.
```
